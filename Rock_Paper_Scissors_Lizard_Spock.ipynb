### To create this game with Python, we need to take the user's choice and then we need to compare it with the computer's choice which is taken using the random module in Python from a list of choices, and if the user wins then the score will increase by +1 if not the computer's score will increase with +1

### The rules of the game are as follows:-

#### Scissors cuts Paper
#### Paper covers Rock
#### Rock crushes Lizard
#### Lizard poizons Spock
#### Spock smashes Scissors
#### Lizard eats Paper
#### Paper disproves Spock
#### Spock vaporises Rock
#### Rock crushes Scissors

```python
import random
choices = ["Rock", "Paper", "Scissor", "Lizard", "Spock"]
computer = random.choice(choices)
palyer = False
cpu_score = 0
player_score = 0

while True:
    player = input("Rock, Papaer, Scissor, Lizard, Spock? ").capitalize()
    ## conditions for the game
    
    if player == computer:
        print("Tie!")
    
    elif player == "Rock":
        if computer == "Paper":
            print("You lose!", computer, "covers", player)
            cpu_score+=1
        elif computer == "Scissor":
            print("You win!", player, "smashes", computer)
            player_score+=1
        elif computer == "Lizard":
            print("You lose!", player, "crushes", computer)
            cpu_score+=1
        else:
            print("You lose!", computer, "vaporises", player)
            cpu_score+=1
    
    
    elif player == "Paper":
        if computer == "Rock":
            print("You win!", player, "covers", computer)
            player_score+=1
        elif computer == "Scissor":
            print("You lose!", computer, "cuts", player)
            cpu_score+=1
        elif computer == "Lizard":
            print("You lose!", computer, "eats", player)
            cpu_score+=1
        else:
            print("You win!", player, "disproves", computer)
            player_score+=1
            
            
            
    
    elif player == "Scissor":
        if computer == "Rock":
            print("You lose!", computer, "crushes", player)
            cpu_score+=1
        elif computer == "Paper":
            print("You win!", player, "cuts", computer)
            player_score+=1
        elif computer == "Lizard":
            print("You win!", player, "decapitates", computer)
            player_score+=1
        else:
            print("You lose!", computer, "smashes", player)
            cpu_score+=1
            
            
    
    elif player == "Lizard":
        if computer == "Rock":
            print("You lose!", computer, "crushes", player)
            cpu_score+=1
        elif computer == "Paper":
            print("You win!", player, "eats", computer)
            player_score+=1
        elif computer == "Scissor":
            print("You lose!", computer, "decapitates", player)
            cpu_score+=1
        else:
            print("You win!", player, "poisons", computer)
            player_score+=1
            
            
    elif player == "Spock":
        if computer == "Rock":
            print("You win!", player, "vaporises", computer)
            player_score+=1
        elif computer == "Paper":
            print("You lose!", computer, "disproves", player)
            cpu_score+=1
        elif computer == "Scissors":
            print("You win!", player, "smashes", computer)
            player_score+=1
        else:
            print("You lose!", computer, "poisons", player)
            cpu_score+=1
            
            
    
    elif player == "End":
        print("Final Scores:")
        print(f"CPU:{cpu_score}")
        print(f"Player:{player_score}")
        break
            
            
            
            
    
```
